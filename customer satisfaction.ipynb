```python
#function to plot pie charts showing percentage of satisfied customers for each flight class
def plot_pie(data, color1, color2, title):
    # Create a list of colors (from iWantHue)
    colors = [color1, color2]

    # Create a pie chart
    plt.pie(
        # using data total)arrests
        data['count'],
        # with the labels being officer names
        labels=data['satisfaction'],
        # with no shadows
        shadow=False,
        # with colors
        colors=colors,
        # with one slide exploded out
        explode=(0, 0.15),
        # with the start angle at 90%
        startangle=90,
        # with the percent listed as a fraction
        autopct='%1.1f%%',
        )

    # View the plot drop above
    plt.axis('equal')
    plt.title(title)
    # View the plot
    plt.tight_layout()
    plt.show()
```

```python
#predictive analytics. 

#import packages
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
import itertools
```

```python
#create dummy variables for categorical variables
data = pd.concat([df,pd.get_dummies(df['Gender'], prefix='Gender'),
                   pd.get_dummies(df['Customer Type'], prefix='Customer Type'),
                  pd.get_dummies(df['Type of Travel'], prefix='Type of Travel'),
                   pd.get_dummies(df['Class'], prefix='Class')],axis=1)
```

```python
#drop dummy encoded data colummns
data = data.drop(['id','Gender','Customer Type','Type of Travel','Class'], axis=1)
```

```python
data.columns
```

```python
#split data into test and train
X = data[['Age', 'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes',
       'Gender_Female', 'Gender_Male', 'Customer Type_Loyal Customer',
       'Customer Type_disloyal Customer', 'Type of Travel_Business travel',
       'Type of Travel_Personal Travel', 'Class_Business', 'Class_Eco',
       'Class_Eco Plus']]
y = data['satisfaction_v2']
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.30, random_state=0)
```

```python
#Initialize logistic regression
logReg = LogisticRegression(max_iter=2000)

#fit logistic regression model on training data
logReg.fit(X_train, y_train)

#predict test data
pred = logReg.predict(X_test)

#check accuracy score
acc = logReg.score(X_test, y_test)
print(acc)
```

```python
#view confusion matrix
cm = metrics.confusion_matrix(y_test, pred)
print(cm)

print('Precision: ', metrics.precision_score(y_test, pred, average='macro'))

print('Recall: ', metrics.recall_score(y_test, pred, average="macro"))

print(classification_report(y_test, pred))
```

```python
#Random Forest

# Create the model with 100 trees
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
# Fit on training data
model.fit(X_train, y_train)
```

```python
#predict with random forest
pre = model.predict(X_test)
```

```python
#print results

print('Accuracy: ', metrics.accuracy_score(y_test, pre))

print(metrics.confusion_matrix(y_test, pre))

print(classification_report(y_test, pre))
```
