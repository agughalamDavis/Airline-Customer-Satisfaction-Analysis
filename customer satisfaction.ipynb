```python
count_eco = pd.DataFrame(eco['satisfaction_v2'].value_counts()).reset_index()
count_eco = count_eco.rename(columns={
    'index':'satisfaction',
    'satisfaction_v2' : 'count'
})

count_business = pd.DataFrame(business['satisfaction_v2'].value_counts()).reset_index()
count_business = count_business.rename(columns={
    'index':'satisfaction',
    'satisfaction_v2' : 'count'
})

count_eco_plus = pd.DataFrame(eco_plus['satisfaction_v2'].value_counts()).reset_index()
count_eco_plus = count_eco_plus.rename(columns={
    'index':'satisfaction',
    'satisfaction_v2' : 'count'
})
count_eco
```

```python
def plot_pie(data, color1, color2, title):
    # Create a list of colors (from iWantHue)
    colors = [color1, color2]

    # Create a pie chart
    plt.pie(
        # using data total)arrests
        data['count'],
        # with the labels being officer names
        labels=data['satisfaction'],
        # with no shadows
        shadow=False,
        # with colors
        colors=colors,
        # with one slide exploded out
        explode=(0, 0.15),
        # with the start angle at 90%
        startangle=90,
        # with the percent listed as a fraction
        autopct='%1.1f%%',
        )

    # View the plot drop above
    plt.axis('equal')
    plt.title(title)
    # View the plot
    plt.tight_layout()
    plt.show()
```

```python
plot_pie(count_eco,'green', 'grey', 'Eco class')
```

```python
plot_pie(count_business,'violet', 'pink', 'Business class')
```

```python
plot_pie(count_eco_plus, 'orange', 'purple','Eco Plus Class')
```
